## Q4. 날짜 문자열

- `end_date`: 오늘 날짜를 `YYYYMMDD` 포맷 문자열로
- `start_date`: 365일 전을 `YYYYMMDD` 포맷 문자열로

````python
# TODO Q4
end_date = datetime.today().strftime("%Y%m%d")
start_date = (datetime.today() - timedelta(days=365)).strftime("%Y%m%d")
````

````python
# CHECK Q4
try:
    assert isinstance(end_date, str) and len(end_date) == 8 and end_date.isdigit(), 'end_date 포맷 오류'
    assert isinstance(start_date, str) and len(start_date) == 8 and start_date.isdigit(), 'start_date 포맷 오류'
    d_end = datetime.strptime(end_date, '%Y%m%d')
    d_start = datetime.strptime(start_date, '%Y%m%d')
    diff = (d_end - d_start).days
    assert 360 <= diff <= 370, f'기간이 1년(±5일)이 아님 (현재: {diff}일)'
    print('✅ PASS Q4')
except Exception as e:
    print(f'❌ FAIL Q4: {e}')
````

<details>
<summary><b>💡 Q4 1차 힌트</b></summary>

- 오늘 객체: `datetime.today()`
- 365일 빼기: `datetime.today() - timedelta(days=365)`
- 문자열로 포맷: `.strftime("%Y%m%d")`

</details>

## Q6. 이동평균 계산 (rolling)

아래 종가 리스트로 DataFrame `ma_df`를 만들고 **MA3**(window=3) 컬럼을 추가하라.

```
종가: [10, 20, 30, 40, 50, 60, 70]
```

````python
# TODO Q6
ma_df = pd.DataFrame({"종가": [10, 20, 30, 40, 50, 60, 70]})
ma_df["MA3"] = ma_df["종가"].rolling(window=3).mean()
````

````python
# CHECK Q6
try:
    assert 'MA3' in ma_df.columns, 'MA3 컬럼 없음'
    assert pd.isna(ma_df['MA3'].iloc[0]), '첫 행 MA3는 NaN이어야 함'
    assert pd.isna(ma_df['MA3'].iloc[1]), '둘째 행 MA3는 NaN이어야 함'
    assert ma_df['MA3'].iloc[2] == 20.0, f'3번째 MA3 기대 20, 실제 {ma_df["MA3"].iloc[2]}'
    assert ma_df['MA3'].iloc[-1] == 60.0, f'마지막 MA3 기대 60, 실제 {ma_df["MA3"].iloc[-1]}'
    print('✅ PASS Q6')
except AssertionError as e:
    print(f'❌ FAIL Q6: {e}')
````

<details>
<summary><b>💡 Q6 1차 힌트</b></summary>

- `df["컬럼"].rolling(window=N).mean()`
- window=3이면 앞 2개 행은 NaN이 되는 게 정상이다.

</details>

<details>
<summary><b>📖 Q6 2차 전체 설명</b></summary>

```python
ma_df = pd.DataFrame({"종가": [10, 20, 30, 40, 50, 60, 70]})
ma_df["MA3"] = ma_df["종가"].rolling(window=3).mean()
```

**이동평균이 하는 일**
- index 2: (10+20+30)/3 = 20
- index 3: (20+30+40)/3 = 30
- ... index 6: (50+60+70)/3 = 60
- 앞의 2개(index 0, 1)는 3개가 안 모여서 NaN.

week1에서 MA5/MA20/MA60 만든 것과 같은 방식.

</details>

<summary><b>📖 Q7 2차 전체 설명</b></summary>

## Q9. 3종목 dict + MA 컬럼 추가

`target_stocks`의 3종목을 모두 조회해 `dfs` dict에 저장하고, 각 DataFrame에 **MA5, MA20, MA60** 컬럼을 추가하라.

구조: `dfs = {'삼성전자': <df>, 'SK하이닉스': <df>, '한미반도체': <df>}`

````python
# TODO Q9
dfs = {}
for name, ticker in target_stocks.items():
    df = stock.get_market_ohlcv_by_date(start_date, end_date, ticker)
    df["MA5"]  = df["종가"].rolling(window=5).mean()
    df["MA20"] = df["종가"].rolling(window=20).mean()
    df["MA60"] = df["종가"].rolling(window=60).mean()
    dfs[name] = df
````

````python
# CHECK Q9
try:
    assert isinstance(dfs, dict), 'dfs가 dict가 아님'
    assert set(dfs.keys()) == {'삼성전자', 'SK하이닉스', '한미반도체'}, f'키 오류: {list(dfs.keys())}'
    for name, df in dfs.items():
        for col in ['MA5', 'MA20', 'MA60']:
            assert col in df.columns, f'{name}: {col} 누락'
        # MA60은 최소 60일 이후부터 값이 있어야 함
        assert df['MA60'].notna().sum() > 0, f'{name}: MA60이 전부 NaN'
    print('✅ PASS Q9')
except AssertionError as e:
    print(f'❌ FAIL Q9: {e}')
````

<details>
<summary><b>💡 Q9 1차 힌트</b></summary>

- `for name, ticker in target_stocks.items()` 로 dict 순회
- 각 티커로 Q8과 같은 방식 조회 → `dfs[name] = df`
- 각 df에 `rolling(window=N).mean()` 3번 (5, 20, 60)

</details>

<details>
<summary><b>📖 Q9 2차 전체 설명</b></summary>

```python
dfs = {}

    df = stock.get_market_ohlcv_by_date(start_date, end_date, ticker)

    df["MA20"] = df["종가"].rolling(window=20).mean()
    df["MA60"] = df["종가"].rolling(window=60).mean()
    dfs[name] = df
```

**흔한 실수**
- 한 df에만 MA 계산하고 `dfs`에 같은 df 3번 넣기 → 세 종목 데이터가 똑같아짐.
- 반드시 반복문 **안에서** 매번 새 df를 받아야 한다.
- MA 컬럼 추가는 `dfs[name] = df` **전**이든 **후**든 관계없다(같은 객체 참조).

## Q12. base64 인코딩 함수

이미지 파일 경로를 받아 base64 문자열을 반환하는 `encode_image(path)` 함수를 만들고, `chart_test.png`를 인코딩해 `img_b64`에 저장하라.

````python
# TODO Q12
def encode_image(path: str) -> str:
    with open(path, "rb") as f:
        return base64.b64encode(f.read()).decode("utf-8")

img_b64 = encode_image("chart_test.png")
````

````python
# CHECK Q12
try:
    assert callable(encode_image), 'encode_image가 함수가 아님'
    assert isinstance(img_b64, str), 'img_b64가 문자열이 아님'
    assert len(img_b64) > 1000, 'base64 문자열이 너무 짧음'
    # 디코딩 검증
    decoded = base64.b64decode(img_b64)
    assert decoded[:8] == b'\x89PNG\r\n\x1a\n', 'PNG 시그니처 불일치'
    print(f'✅ PASS Q12 (길이 {len(img_b64)})')
except Exception as e:
    print(f'❌ FAIL Q12: {e}')
````

<details>
<summary><b>💡 Q12 1차 힌트</b></summary>

- 파일을 **바이너리 모드**로 열기: `open(path, "rb")`
- `base64.b64encode(bytes)` → bytes 반환 → `.decode("utf-8")`로 문자열 변환

</details>

<details>
<summary><b>📖 Q12 2차 전체 설명</b></summary>


**왜 base64인가**
- OpenAI API는 JSON으로 요청을 받는다. JSON에 이진 바이트를 직접 못 넣는다.
- base64는 이진 → ASCII 문자열 변환 방식. 크기가 약 33% 커지지만 JSON에 안전하게 실린다.
- 나중에 `"data:image/png;base64,{img_b64}"` 형태로 API에 전달.

</details>

## Q13. Vision API 호출

`client.chat.completions.create`를 사용해 **gpt-4o** 모델에 차트 이미지를 보내 분석 결과를 `ai_analysis`에 저장하라.

요구:
- 모델: `gpt-4o`
- user message에 `text`와 `image_url` 둘 다 포함
- `image_url`은 `data:image/png;base64,{img_b64}` 형식

````python
# TODO Q13
response = client.chat.completions.create(
    model="gpt-4o",
    messages=[
        {"role": "user", "content": [
            {"type": "text",
             "text": "이 차트는 삼성전자의 최근 1년 일봉이다. 이평선 배열을 중심으로 분석하라."},
            {"type": "image_url",
             "image_url": {"url": f"data:image/png;base64,{img_b64}"}},
        ]},
    ],
    max_tokens=500,
    temperature=0.3,
)
ai_analysis = response.choices[0].message.content
print(ai_analysis)
````
